**Pass**/**Fail** **Analyzer**

```python
marks=[45,78,90,33,60]
pass_count = 0
fail_count = 0
for mark in marks:
    if mark >= 50:
      pass_count +=1

    else:
      fail_count +=1
print("Total number of pass students is : ",pass_count)
print("Total number of fail students is : ",fail_count)


```

**Simple Data Cleaner**

```python
names=[" Alice ", "bob", "CHARLIE"]
cleaned_names=[]
for name in names:
  cleaned_name=name.strip().lower()
  cleaned_names.append(cleaned_name)
print(cleaned_names)
```

**Message Length Analyzer**

```python
messages=["Hi","Welcome to the platform","OK"]
for msg in messages:
  length=len(msg)
  print(f"Message : '{msg}'->Length:{length}")

  if length>10:
    print("This message is longer than 10 characters")

```
