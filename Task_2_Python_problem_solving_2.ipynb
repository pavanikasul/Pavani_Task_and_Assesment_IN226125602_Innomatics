**1. Unique Words in a Sentence**

```python
text=input("Enter a Sentence: ")
words=text.split()

unique_words=[]

for word in words:
  if word not in unique_words:
    unique_words.append(word)

print(unique_words)
print("count of unique_words:",len(unique_words))
```

**3.Find Maximum and Minimum Values**


```python
numbers=[45,22,89,10,66]
max_number=numbers[0]
min_number=numbers[0]
for number in numbers:
  if number > max_number:
    max_number=number
  if number < min_number:
    min_number=number
print("List:",numbers)
print("Maximum number is : ",max_number)
print("Minimum number is : ",min_number)

```

**4.Count Products Above a Price Threshold**


```python
prices=[450, 1200, 899, 1500, 300]
count=0
for price in prices:
  if price > 1000:
    count=count+1
  else:
    count=count

print("Products above 1000 :",count)
```

**7.Count Character Frequency**

```python
text = "ppython"
char_count = {}
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)

```
