**1.Employee Performance Bonus Eligibility**

```python
employees = {
    "Ravi":92,
    "Anita":88,
    "Kiran":92,
    "Suresh":85
}
highest_score = max(employees.values())

top_performers = []
for name, score in employees.items():
  if score == highest_score:
    top_performers.append(name)
print("Top Performers Eligible for Bonus:", ",".join(top_performers),
      f"(Score: {highest_score})")
```

**Search Query Keyword Analysis**

```python
text=input("Enter the sentence :")
text = text.lower()
print("The text is converted to lowe case :",text)
words = text.lower().split()
frequency = {}
for word in words:
  if word in frequency:

    frequency[word]+=1
  else:
    frequency[word]=1

result ={}
for word,count in frequency.items():
  if count >1:
    result[word]=count
print(result)




```

**3.Sensor Data Validation**

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
reading_value = []
for index, value in enumerate(sensor_readings):
  if value %2 ==0:
    reading_value.append((index,value))
print("Valid Sensor Readings (Hour, Value):")
print(reading_value)

```

**4.Email Domain Usage Analysis**

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count = {}
for email in emails:
  user=email.split("@")[1]
  if user in domain_count:
    domain_count[user]+=1
  else:
    domain_count[user]=1
total_emails=len(emails)
for user,count in domain_count.items():
  percentage=(count/total_emails)*100
  print(f"{user}:{int(percentage)}%")

```

**5.Sales Spike Detection**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
daily_average = sum(sales)/len(sales)
print("Daily Average sales:", daily_average)

threshold = daily_average *1.30

for day, value in enumerate(sales):
  if value > threshold:
    print(f"Day{day+1}:{value}")

```

**6.Duplicate User ID Detection**

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
users = {}
for user in user_ids:
  if user in users:
    users[user]+=1
  else:
    users[user]=1
print(users)

for user,count in users.items():
  if count>1:
    print(f"{user}->{count}times")
```
