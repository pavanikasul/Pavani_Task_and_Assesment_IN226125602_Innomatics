**1.Mobile Recharge Validation System**

```python
def mobile_recharge(recharge_amount):
  valid_plans=[ 199, 299, 399, 599]

  if recharge_amount < 50:
    return "The amount must be at least 50"
  if recharge_amount not in valid_plans:
    return "Invalid plan selected"
  return "Recharge Successfull"
while True:
  recharge_amount=int(input("Enter Recharge amount: "))

  result=mobile_recharge(recharge_amount)
  print(result)

  if result == "Recharge Successfull":
    break
  else:
    print("Please try again!")
```

**2.Inventory Reorder Alert System**

```python
products={
    "MoongDal":15,
    "AlooBhuja":30,
    "kurkure":10,
    "JimJam":50,
    "DailryMilk-silk":100
}
def reorder(products):
  for product , stock in products.items():
    if stock < 15:
      print(product,"-Reorder Alert")
    else:
      print(product,"-Stock Ok")
reorder(products)




```

**3.Student Result Processing System**

```python
marks=[33,50,89,78,99,63,20,10]

def average_marks(marks):
  average_marks=sum(marks)/len(marks)
  print(average_marks)
  if average_marks>=50:
    return "Pass"
  else:
    return "Fail"
result=average_marks(marks)
print(result)
```

**4.Cab Fare Estimator with Retry Option**

```python
def cab_calculation(distance,peak):
  base_fare=50
  rate=12
  fare=base_fare+(distance * rate)
  if peak =="yes":
    fare+=fare*0.25
  return fare
while True:
  distance=float(input("Enter distance in km: "))
  peak=input("Is it peak hour? (yes/no):").lower()

  total_fare=cab_calculation(distance,peak)
  print("Total Fare: ",total_fare)

  retry = input("Do you want to calculate again?(yes/no):").lower()
  if retry!="yes":
    print("Thank you!")
    break
```

**5.Employee Attendance Eligibility**

```python
attendance=["p","A"]
def attendance_percentage(attendance):
  present_days=0
  for day in attendance:
    if day =="p":
      present_days+=1
  attendance_percentage=(present_days/len(attendance))*100
  print("Attendance percentage:", attendance_percentage)
  if attendance_percentage >= 75:
    return "Eligible"
  else:
    return "Not Eligible "

result=attendance_percentage(attendance)
print(result)


```

**6.Password strength checker**

```python
def check_password(password):
  has_digit=False
  has_special=False

  if len(password)<8:
    return "The length of password should be at least 8 characters"
  for char in password:
    if char.isdigit():
      has_digit=True
    if char in "@#$":
      has_special=True
  if not has_digit:
    return "The password should contain at least one digit"
  if not has_special:
    return "The password should contain at leat one speacial character"

  return "Login Successful"
password=(input("Enter your password: "))
result=check_password(password)
print(result)

```
